# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:    
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return bmi  

    
````

````python
# Create a function that converts temperature from Celsius to Fahrenheit
# Include input validation and round to 2 decimal places

# Try typing this comment and let Copilot suggest the implementation
def celsius_to_fahrenheit(celsius):
    # Convert Celsius to Fahrenheit
    if celsius < -273.15:
        raise ValueError("Temperature below absolute zero is not possible.")
    fahrenheit = (celsius * 9/5) + 32
    return round(fahrenheit, 2)
# Test the function with a few examples
print(celsius_to_fahrenheit(0))  # Expected output: 32.0
print(celsius_to_fahrenheit(100))  # Expected output: 212.0
print(celsius_to_fahrenheit(-40))  # Expected output: -40.0

````

## 6. Test Generation

Copilot can help write tests for your code:

````python
import pytest

def calculate_dosage(weight, age, condition):
    """Calculate medication dosage based on patient parameters."""
    if weight <= 0 or age < 0:
        raise ValueError("Weight and age must be positive")
    
    base_dosage = weight * 2
    if age < 12:
        base_dosage *= 0.75
    
    if condition == "severe":
        base_dosage *= 1.5
    
    return round(base_dosage, 2)

# Write pytest test cases for the above function
# Import the pytest module for testing

# Test normal case: weight = 50, age = 30, condition = "normal"
def test_normal_case():
    assert calculate_dosage(50, 30, "normal") == 100.0

# Test case for a child: weight = 20, age = 10, condition = "normal"
def test_child_case():
    assert calculate_dosage(20, 10, "normal") == 30.0

# Test case for severe condition: weight = 50, age = 30, condition = "severe"
def test_severe_condition():
    assert calculate_dosage(50, 30, "severe") == 150.0

# Test edge case: weight = 0, age = 30, condition = "normal" (should raise ValueError)
def test_zero_weight():
    with pytest.raises(ValueError):
        calculate_dosage(0, 30, "normal")

# Test edge case: weight = 50, age = -1, condition = "normal" (should raise ValueError)
def test_negative_age():
    with pytest.raises(ValueError):
        calculate_dosage(50, -1, "normal")

# Test edge case: weight = 50, age = 10, condition = "severe"
def test_child_severe_condition():
    assert calculate_dosage(50, 10, "severe") == 112.5





````

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Process a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    if image.size == 0:
        raise ValueError("Input image cannot be empty.")
    
    # Normalize the image to the range [0, 1]
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))
    
    # Apply a Gaussian filter for smoothing
    filtered_image = gaussian_filter(normalized_image, sigma=1)
    
    return filtered_image

# Example usage
if __name__ == "__main__":
    # Create a sample image (e.g., a 2D numpy array)
    sample_image = np.random.rand(100, 100) * 255  # Simulating a grayscale image
    processed_image = process_medical_image(sample_image)
    print("Processed image shape:", processed_image.shape)

    import numpy as np
import matplotlib.pyplot as plt

# Create a sample image (e.g., a 2D numpy array simulating a grayscale image)
sample_image = np.random.rand(100, 100) * 255  # Random pixel values between 0 and 255

# Process the image using the function
processed_image = process_medical_image(sample_image)

# Visualize the original and processed images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(sample_image, cmap="gray")
plt.colorbar()

# Processed image
plt.subplot(1, 2, 2)
plt.title("Processed Image")
plt.imshow(processed_image, cmap="gray")
plt.colorbar()

plt.tight_layout()
plt.show()
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    def __init__(self):
        """Initialize an empty dictionary to store patient records."""
        self.records = {}

    def add_record(self, patient_id, name, age, weight, condition):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Patient's name.
            age (int): Patient's age.
            weight (float): Patient's weight in kilograms.
            condition (str): Patient's medical condition.

        Raises:
            ValueError: If any input is invalid.
        """
        if not patient_id or not isinstance(patient_id, str):
            raise ValueError("Patient ID must be a non-empty string.")
        if not name or not isinstance(name, str):
            raise ValueError("Name must be a non-empty string.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if not isinstance(weight, (int, float)) or weight <= 0:
            raise ValueError("Weight must be a positive number.")
        if not condition or not isinstance(condition, str):
            raise ValueError("Condition must be a non-empty string.")

        self.records[patient_id] = {
            "name": name,
            "age": age,
            "weight": weight,
            "condition": condition,
        }

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            kwargs: Fields to update (e.g., name, age, weight, condition).

        Raises:
            KeyError: If the patient ID does not exist.
            ValueError: If any input is invalid.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")

        for key, value in kwargs.items():
            if key not in self.records[patient_id]:
                raise ValueError(f"Invalid field: {key}")
            if key == "age" and (not isinstance(value, int) or value <= 0):
                raise ValueError("Age must be a positive integer.")
            if key == "weight" and (not isinstance(value, (int, float)) or value <= 0):
                raise ValueError("Weight must be a positive number.")
            self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient record.

        Raises:
            KeyError: If the patient ID does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")
        return self.records[patient_id]

    def calculate_average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: Average age.

        Raises:
            ValueError: If no records exist.
        """
        if not self.records:
            raise ValueError("No records available to calculate average age.")
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def calculate_average_weight(self):
        """
        Calculate the average weight of all patients.

        Returns:
            float: Average weight.

        Raises:
            ValueError: If no records exist.
        """
        if not self.records:
            raise ValueError("No records available to calculate average weight.")
        total_weight = sum(record["weight"] for record in self.records.values())
        return total_weight / len(self.records)

# Example usage
if __name__ == "__main__":
    records = PatientRecords()
    records.add_record("P001", "Alice", 30, 65.5, "Normal")
    records.add_record("P002", "Bob", 45, 80.0, "Diabetes")
    records.update_record("P001", weight=66.0)
    print("Record for P001:", records.get_record("P001"))
    print("Average age:", records.calculate_average_age())
    print("Average weight:", records.calculate_average_weight())

# Example of error handling for emprty records
try:
    records.calculate_average_age()
except ValueError as e:
    print("Error:", e)
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Sample dataset
data = {
    "Patient ID": ["P001", "P002", "P003", "P004", "P005"],
    "Age": [30, 45, 25, 60, 50],
    "Weight": [65.5, 80.0, 55.0, 90.0, 75.0],
    "Condition": ["Normal", "Diabetes", "Normal", "Hypertension", "Diabetes"],
}
df = pd.DataFrame(data)

# Function to display a bar chart of patient ages
def plot_age_distribution(condition_filter="All"):
    plt.figure(figsize=(8, 5))
    if condition_filter == "All":
        filtered_df = df
    else:
        filtered_df = df[df["Condition"] == condition_filter]
    
    plt.bar(filtered_df["Patient ID"], filtered_df["Age"], color="skyblue")
    plt.title(f"Age Distribution ({condition_filter})")
    plt.xlabel("Patient ID")
    plt.ylabel("Age")
    plt.xticks(rotation=45)
    plt.show()

# Function to display a scatter plot of weight vs. age
def plot_weight_vs_age():
    plt.figure(figsize=(8, 5))
    plt.scatter(df["Age"], df["Weight"], color="green", alpha=0.7)
    for i, patient_id in enumerate(df["Patient ID"]):
        plt.text(df["Age"][i], df["Weight"][i], patient_id, fontsize=9)
    plt.title("Weight vs. Age")
    plt.xlabel("Age")
    plt.ylabel("Weight")
    plt.grid(True)
    plt.show()

# Interactive dashboard
def dashboard(condition_filter="All"):
    print("Bar Chart: Age Distribution")
    plot_age_distribution(condition_filter)
    print("\nScatter Plot: Weight vs. Age")
    plot_weight_vs_age()

# Create an interactive widget for filtering by condition
interact(
    dashboard,
    condition_filter=widgets.Dropdown(
        options=["All"] + df["Condition"].unique().tolist(),
        value="All",
        description="Condition:",
    ),
)
````
